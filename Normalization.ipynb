```python
from matplotlib import pyplot
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

## Define Normalization Function

```python
def normalize(data):
    y = pd.to_numeric(data)
    y = np.array(y.reshape(-1, 1))
    
    scaler = MinMaxScaler()
    y = scaler.fit_transform(y)
    y = y.reshape(1, -1)[0]
    return y
```

## Read Dataset (Flood)

```python
# Read the dataset
ROAD = "Taft Ave."
YEAR = "2015"
EXT = ".csv"
#FILENAME = "merged_mmda_wwo_" + ROAD + "_" + YEAR
FILENAME = "interpolated_flood_pagasa"
FOLDER = "flood"
data = pd.read_csv("data/" + FOLDER + "/" + FILENAME + EXT, skipinitialspace=True)
```

```python
data['WL [El.m]'] = normalize(data['WL [El.m]'])
data
```

## Export new Dataset with Normalized Data

```python
data.to_csv("data/" + FOLDER + "/" + FILENAME + EXT, encoding='utf-8', index=False)
```

```python
startIndex = 0
endIndex = (24*4)*365 #If not Leap Year
#endIndex = (24*4)*365+1 #If Leap Year
newdata = data.iloc[startIndex:endIndex, :]

YEAR = "2015"
newdata.to_csv("data/" + FOLDER + "/" + FILENAME + "_" + YEAR + EXT, encoding='utf-8', index=False)
```
